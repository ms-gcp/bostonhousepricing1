```python
### dependent & independent features
X = df.iloc[:,:-1]   # skip the last column
y = df.iloc[:,-1]    # consider only last column
```

```python
X.head()
```

```python
y
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

### Standardizing the Dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train) # transform it completely 
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## Print tha coefficient 
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## Prediction with test data (prediction on bulk or batch data)
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

### Assumptions

```python
## Plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
## Residual (difference between actual & predicted value)
## it shows normal distribution means mosdel is performing well
residuals =y_test-reg_pred
residuals
```

```python
## Plotting this residuals
sns.displot(residuals,kind='kde')
```

```python
## scatter plotwith respect to prediction & residuals
## Unifromly distribution of predicted values & resuduals shows model performing well
plt.scatter(reg_pred,residuals)
```

### Performance metrics

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```
